```python
def findFr(frMean):
    diff =  np.abs(fr - frMean)
    diff_mean = np.mean(diff,axis=(1,2))
    diff_mean.shape
    mean_index = np.argsort(diff_mean)[:100]
    _ , axs =  plt.subplots(10,10,figsize=(10,10))
    cnt = 0
    for i in range(10):
        for j in range(10):
            axs[i,j].imshow(fr[mean_index[cnt]],cmap='gray_r')
            axs[i,j].axis('off')
            cnt += 1
    plt.show()
```

```python
findFr(appleMean)
```

```python
findFr(pineAppleMean)
```

```python
findFr(bananaMean)
```

```python
from sklearn.cluster import KMeans
```

```python
fr = np.load('fruits_300.npy')
```

```python
km = KMeans(n_clusters=3,random_state=42)
```

```python
fr2 = fr.reshape(300,-1)
km.fit(fr2)
```

```python
help(KMeans)
```

```python
km.labels_
```

```python
np.unique(km.labels_, return_counts=True)
```

```python
cluster2_img = fr[km.labels_ == 2]
```

```python
cluster2_img.shape[0] # 91
for i in range(10):
    for j in range(10):
        pass
    
# plt.imshow(cluster2_img[0],cmap='gray_r')
```

```python
def showCluster(cluster_img):    
    _ , axs =  plt.subplots(10,10,figsize=(10,10))
    cnt = 0
    for i in range(10):
        for j in range(10):
            if cnt < 91:
                axs[i,j].imshow(cluster_img[cnt],cmap='gray_r')
            axs[i,j].axis('off')
            cnt += 1            
    plt.show()
```

```python
cluster2_img = fr[km.labels_ == 2]
showCluster(cluster2_img)
```

```python
cluster2_img = fr[km.labels_ == 1]
showCluster(cluster2_img)
```

```python
cluster2_img = fr[km.labels_ == 0]
showCluster(cluster2_img)
```

```python
km.cluster_centers_.shape
clusterCenter =  km.cluster_centers_.reshape(3,100,100)
clusterCenter.shape
```

```python
_,axs =  plt.subplots(1,3,figsize=(20,10))
axs[0].imshow(clusterCenter[0], cmap ='gray_r')
axs[0].axis('off')
axs[1].imshow(clusterCenter[1], cmap='gray_r')
axs[1].axis('off')
axs[2].imshow(clusterCenter[2], cmap='gray_r')
axs[2].axis('off')
plt.show()
```
