```python
# matplotlib  맷플롯립
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 그림(figure)
# 축(axes)
x = range(100)
y = range(100)
plt.plot(x,y)
plt.show()
```

```python
x_1 = range(100)
y_1 = [np.cos(i) for i in x]

x_2 = range(100)
y_2 = [np.sin(i) for i in x]

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.show()
```

```python
# 축에대한 개념 및 사용
fig, ax = plt.subplots()  # figure 와 axes객체 할당
x_1 = range(100)
y_1 = [np.cos(i) for i in x]

ax.plot(x_1,y_1)
ax.set(title='cos graphic', xlabel='x', ylabel='y')
plt.show()
```

```python
fig = plt.figure() #figure 객체 반환
fig.set_size_inches(10,5)

ax_1 =  fig.add_subplot(1,2,1)
ax_2 =  fig.add_subplot(1,2,2)

ax_1.plot(x_1,y_1, c='b')
ax_2.plot(x_2,y_2, c='g')
plt.show()
```

```python
x = np.linspace(-1,1,100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
plt.show()
```

```python
x = range(100)
y = range(100)

x2 = range(100)
y2 = [i+10 for i in x2]

x3 = range(100)
y3 = [i+20 for i in x3]

x4 = range(100)
y4 = [i+30 for i in x4]

plt.plot(x,y, color = "#000000")
plt.plot(x2,y2, c="b")
plt.plot(x3,y3, c="r", linestyle='dashed')
plt.plot(x4,y4, c="g", linestyle='dotted')
plt.title("varias line style")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#막대그래프 : 데이터의 개수나 크기를 비교
from random import sample
# 1. 데이터생성
data = [
    sample(range(100),5),
    sample(range(100),5),
    sample(range(100),5)
]
# 2. x좌표 시작점
x = np.arange(0,10,2)

# 3개의 막대그래프  생성
plt.bar(x,data[0],color='b',width=0.50)
plt.bar(x+0.50,data[1],color='g',width=0.50)
plt.bar(x+1.0,data[2],color='r',width=0.50)

# 4. x축에 표시될 이름과 위치 설정
plt.xticks(x+0.50, ('A','B','C','D','E'))

# 5. 출력
plt.show()
```

```python
# 누적막대그래프
data = np.array([
    sample(range(100),5),
    sample(range(100),5),
    sample(range(100),5)
])
color_list = ['b','g','r']
data_label = ['A','B','C']
x = np.arange(data.shape[1])
for i in range(3):
    plt.bar(x,data[i],
           bottom = np.sum(data[:i], axis=0),
            color = color_list[i],
            label=data_label[i]
           )
plt.legend()
plt.show()
```

```python
# 히스토그램  : 데이터의 분포를 표현( 빈도)
n = 1000
x = np.random.normal(size=n)
plt.hist(x,bins=100)
plt.show()
```

```python
# 시본 seaborn : 맷플롯플립을 Wrapper 모듈
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
sns.set_style("whitegrid")
sns.lineplot(x="timepoint", y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y='signal',data=fmri, hue = 'event')
plt.show()
```

```python
# 회귀 그래프 : 회귀식을 적용해서 선형회귀 추세선을 그래프에 함께 작성
# 추세선 : x값대비 y값변화를 예측하는 직선
# regplot 함수 사용 
```

```python
tips =  sns.load_dataset("tips")
tips.head()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips,ci = 95) # ci 신뢰구간의 비율
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, ci=50) # ci 신뢰구간
plt.show()
```

```python
# 산점도( scatter plot)  x y를 기준으로 데이터의 분포
sns.scatterplot(x='total_bill', y='tip',data=tips,hue='time')
plt.show()
```

```python
# count plot : 범주형 데이터의 항목별 개수
```

```python
sns.countplot(x="day", hue="time", data=tips)
plt.show()
```

```python
# 막대그래프    y값이 연속형 값이 경우 해당 값들의 평균
# 데이터의 신뢰구간을 검은색 막대로 표현
sns.barplot(x="day", y='total_bill', data=tips)
plt.show()
```

```python
# 여러그래프들을 합쳐서 정보를 추출 - 특히 범주형 데이터에 유횽
# 바이올린 플롯 
#  - 상자그림 + 분포도
# x : 범주형  y : 연속형

# 스웜플롯
# - 바이올린 플롯  + 산점도
```

```python
sns.violinplot(x='day',y='total_bill', hue='smoker', data=tips, palette="muted")
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill', hue='smoker', data=tips, palette="muted")
plt.show()
```

```python
# 다양한 범주형 데이터를 나타내는 패싯그리드
# FacetGrid
```

```python
g = sns.FacetGrid(tips,col="time", row='sex')
g.map(sns.scatterplot,"total_bill",'tip')
plt.show()
```

```python
g = sns.FacetGrid(tips,col="time", row='sex')
g.map_dataframe(sns.histplot,x='total_bill')
plt.show()
```
