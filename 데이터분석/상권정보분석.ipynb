```python
#시군구명 별 상권업종중분류명 별 업종수
df_food_group =  pd.DataFrame( df_food.groupby(['시군구명','상권업종중분류명'])['상권업종중분류명'].count())
# df_food_group.loc['강남구']
df_food_group.columns = ['상권업종분류갯수']
# df_food_group.reset_index(inplace=True)
df_food_group
```

```python
city_lists =  df_food['시군구명'].unique()
len(city_lists), city_lists
```

```python
fig = plt.figure(figsize=(15,5))

def showSns(location:str, idx:int)->None:    
    ax = fig.add_subplot(5,5,idx)
    temp = df_food_group.loc[location].sort_values(by = '상권업종분류갯수',ascending=False)
    ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
    ax.set_title(location)
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(5,5,1)
temp = df_food_group.loc['서초구'].sort_values(by = '상권업종분류갯수',ascending=False)
ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
ax.set_title('서초구')
ax = fig.add_subplot(5,5,2)
temp = df_food_group.loc['동작구'].sort_values(by = '상권업종분류갯수',ascending=False)
ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
ax.set_title('동작구')
ax = fig.add_subplot(5,5,3)
temp = df_food_group.loc['강남구'].sort_values(by = '상권업종분류갯수',ascending=False)
ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
ax.set_title('강남구')
ax = fig.add_subplot(5,5,4)
temp = df_food_group.loc['강동구'].sort_values(by = '상권업종분류갯수',ascending=False)
ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
ax.set_title('강동구')
ax = fig.add_subplot(5,5,5)
temp = df_food_group.loc['마포구'].sort_values(by = '상권업종분류갯수',ascending=False)
ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
ax.set_title('마포구')
ax = fig.add_subplot(5,5,6)
temp = df_food_group.loc['송파구'].sort_values(by = '상권업종분류갯수',ascending=False)
ax = sns.barplot(data=temp, x=temp.index,y = temp['상권업종분류갯수'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 80)
ax.set_title('송파구')
```
