```python
# matplotlib  맷플롯립
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 그림(figure)
# 축(axes)
x = range(100)
y = range(100)
plt.plot(x,y)
plt.show()
```

```python
x_1 = range(100)
y_1 = [np.cos(i) for i in x]

x_2 = range(100)
y_2 = [np.sin(i) for i in x]

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.show()
```

```python
# 축에대한 개념 및 사용
fig, ax = plt.subplots()  # figure 와 axes객체 할당
x_1 = range(100)
y_1 = [np.cos(i) for i in x]

ax.plot(x_1,y_1)
ax.set(title='cos graphic', xlabel='x', ylabel='y')
plt.show()
```

```python
fig = plt.figure() #figure 객체 반환
fig.set_size_inches(10,5)

ax_1 =  fig.add_subplot(1,2,1)
ax_2 =  fig.add_subplot(1,2,2)

ax_1.plot(x_1,y_1, c='b')
ax_2.plot(x_2,y_2, c='g')
plt.show()
```

```python
# 서브플롯 행렬
fig, ax =  plt.subplots(nrows=1, ncols=2)
print(ax)
```

```python
x = np.linspace(-1,1,100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
plt.show()
```

```python
x = range(100)
y = range(100)

x2 = range(100)
y2 = [i+10 for i in x2]

x3 = range(100)
y3 = [i+20 for i in x3]

x4 = range(100)
y4 = [i+30 for i in x4]

plt.plot(x,y, color = "#000000")
plt.plot(x2,y2, c="b")
plt.plot(x3,y3, c="r", linestyle='dashed')
plt.plot(x4,y4, c="g", linestyle='dotted')
plt.title("varias line style")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 4개의 화면 2 x 2 화면에 위의 선과 종류를 각각 그려보세요.
fig =  plt.figure()
fig.set_size_inches(10,10)
ax_1 = fig.add_subplot(2,2,1)
ax_2 = fig.add_subplot(2,2,2)
ax_3 = fig.add_subplot(2,2,3)
ax_4 = fig.add_subplot(2,2,4)

x = range(100)
y = range(100)

ax_1.plot(x,y,color = "#000000")
ax_2.plot(x,y,c='b')
ax_3.plot(x,y,c='r', linestyle = 'dashed')
ax_4.plot(x,y,c='g', linestyle='dotted')
plt.show()
```
