```python
# OR 데이터에 퍼셉트론을 적용
from sklearn.linear_model import Perceptron
```

```python
# 훈련데이터를 생성
x = [[0,0],[0,1],[1,1],[1,0]]
y = [-1,1,1,1]
```

```python
# 학습
p =  Perceptron()
p.fit(x,y)
```

```python
print(f"학습된 퍼셉트론의 매개변수 = 기울기:{p.coef_}, 절편:{p.intercept_} ")
```

```python
print(f"훈련집합에 대한 예측 : {p.predict(x)}")
```

```python
print(f"정확률 = {p.score(x,y)*100}%")
```

```python
# 필기숫자 데이터에 퍼셉트론을 적용
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
digit = datasets.load_digits()
x_train,x_test,y_train,y_test =   train_test_split(digit.data, digit.target,random_state=42)
```

```python
# 퍼셉트론 적용
p = Perceptron(max_iter=100)
```

```python
p.fit(x_train,y_train)
```

```python
predict = p.predict(x_test)
predict, y_test
```

```python
# predict, y_test
p.score(x_test,y_test)
```

```python
# 혼동행열
conf = np.zeros((10,10))
for i in range(len(predict)):
    conf[predict[i],y_test[i]] += 1
conf    
```

```python
np.sum([conf[i,i] for i in range(10)]) / len(predict)
```

```python
# svm 비선형 분류기
from sklearn import svm
```

```python
ss = svm.SVC()
ss.fit(x_train,y_train)
predict = ss.predict(x_test)
ss.score(x_test,y_test)
```
