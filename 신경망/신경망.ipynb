```python
# 퍼셉트론의 학습
# 실제상황은 데이터만 주어지지만, 학습 알고리즘으로 가중치를 알아내야함
# sklearn의 필기숫자  64차원의 특징벡터 1797개의 샘플 - >학습알고리즘 없이 불가능
```

```python
# 동작을 취한다.
# while(true):
#     수영동작을 하고 평가
#     if 만족:
#         break
#     더 개선된 방향으로 동작을  수정
# 동작을 저장    
```

```python
# 초기동작 w = (팔을돌린다, 팜꿈치 각도, 팔과 귀의 거리) 초기화
# while(true):
#     w 에 따라서 수영을 하고 동작 w의 점수 j(w)를 계산
#     if(j(w) 만족스러움) break
#     손실함수값을 낮추는 방향 W(improve)를 계산
#     w = w + W(inprove)
# w를 저장    
```

```python
# 신경망 학습
# 신경망 학습- 최적화의 문제. 최적화의대상- 신경망의 가중치(매개변수)
# 손실함수 - 오류의 양을 측정
# 낮추는 방향 W(improve) - 미분을 사용
```

```python
# 손실함수 j를 설계
# w가 훈련집합에 있는 샘플을 모두 맞추면 즉 정확율이 100%이면 j(w)는 0
# w가 틀리는 샘플이 많을수록 j(w)값은 크다
```
